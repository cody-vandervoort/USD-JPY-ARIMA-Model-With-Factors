```python
import pandas as pd
```

```python
x1 = pd.read_csv('xrp.csv')
x1.columns = ['ds', 'OPEN', 'HIGH', 'LOW', 'yhat',]
x1['ds']= pd.to_datetime(x1['ds'])
x1['ds'].dt.tz_localize(None),
x1["ds"] = x1["ds"].dt.strftime("%Y-%m-%d %H:%M:%S")
x1 = x1.set_index(pd.DatetimeIndex(x1['ds']))
```

```python
del x1['OPEN']
del x1['HIGH']
del x1['LOW']
del x1['ds']
x1.columns = ['xrp']
x1 = x1.iloc[-500:]
x1.to_csv('xrp.csv')
```
